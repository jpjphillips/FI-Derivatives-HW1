```python
import pandas as pd
October_10 = pd.read_excel('option_data_bb_SFRZ5.xlsx', sheet_name='2025-10-10')
November_14 = pd.read_excel('option_data_bb_SFRZ5.xlsx', sheet_name='2025-11-14')
spot = pd.read_excel('option_data_bb_SFRZ5.xlsx', sheet_name='spot')
```

```python
import numpy as np
from scipy.stats import norm

# ===== Inputs =====
F0 = 96.085           # futures price directly (NOT spot -> forward)
K = 96.0625
r = 0.04396
sigma = 0.1854 *(100 - F0)/100    # your implied vol
T = 73 / 365
market_price = 0.1300

# ===== Black (futures) =====
sigma_sqrt_T = sigma * np.sqrt(T)

d1 = (np.log(F0 / K) + 0.5 * sigma**2 * T) / sigma_sqrt_T
d2 = d1 - sigma_sqrt_T

discount_factor = np.exp(-r * T)

call_price = discount_factor * (
    F0 * norm.cdf(d1) - K * norm.cdf(d2)
)

print("Black model price:", round(call_price, 6))
print("Market price:", market_price)
print("Difference:", round(call_price - market_price, 6))

```

```python
call_prices = pd.DataFrame(columns=['strike price', 'call price', 'market price', 'difference', 'implied vol rates', 'implied vol price'])
put_prices = pd.DataFrame(columns=['strike price', 'put price', 'market price', 'difference', 'implied vol rates', 'implied vol price'])

for i, row in October_10.iterrows():
    F0 = 96.085
    K = row['strike price']

    r = 0.04396
    sigma = row['implied vol'] *(100 - F0)/100
    T = 73 / 365
    sigma_sqrt_T = sigma * np.sqrt(T)

    d1 = (np.log(F0 / K) + 0.5 * sigma**2 * T) / sigma_sqrt_T
    d2 = d1 - sigma_sqrt_T

    discount_factor = np.exp(-r * T)
    market_price = row['price']

    if row['option type'] == 'C':


        call_price = discount_factor * (
            F0 * norm.cdf(d1) - K * norm.cdf(d2)
        )

        print(f"Strike: {K}, Black model price: {round(call_price, 6)}, Market price: {market_price}, Difference: {round(call_price - market_price, 6)}")
        call_prices.loc[len(call_prices)] = {'strike price': K, 'call price': round(call_price, 6), 'market price': market_price, 'difference': round(call_price - market_price, 6), 'implied vol rates': row['implied vol'], 'implied vol price': round(sigma, 6)}

    else:
        put_price = discount_factor * (
            K * norm.cdf(-d2) - F0 * norm.cdf(-d1)
        )

        print(f"Strike: {K}, Black model price: {round(put_price, 6)}, Market price: {market_price}, Difference: {round(put_price - market_price, 6)}")
        put_prices.loc[len(put_prices)] = {'strike price': K, 'put price': round(put_price, 6), 'market price': market_price, 'difference': round(put_price - market_price, 6), 'implied vol rates': row['implied vol'], 'implied vol price': round(sigma, 6)}
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(18, 6))
plt.subplot(1, 2, 1)
plt.plot(call_prices['strike price'], call_prices['call price'], label='Black Model Call Price', marker='o')
plt.plot(call_prices['strike price'], call_prices['market price'], label='Market Call Price', marker='x')
plt.plot(put_prices['strike price'], put_prices['put price'], label='Black Model Put Price', marker='o')
plt.plot(put_prices['strike price'], put_prices['market price'], label='Market Put Price', marker='x')
plt.title('Call Option Prices')
plt.xlabel('Strike Price')
plt.ylabel('Price')
plt.legend()
```

Well, the Market believes that the optionality is less valuable than the model, maybe this is because it is not used optimally.

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(call_prices['strike price'], call_prices['implied vol price'], label='Calls', marker='o')
plt.plot(put_prices['strike price'], put_prices['implied vol price'], label='Puts', marker='x')
plt.title('Implied Volatility Prices')
plt.xlabel('Strike Price')
plt.ylabel('Implied Volatility Price')
plt.legend()

plt.show()
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(call_prices['strike price'], call_prices['implied vol rates'], label='Calls', marker='o')
plt.plot(put_prices['strike price'], put_prices['implied vol rates'], label='Puts', marker='x')
plt.title('Implied Volatility Rates')
plt.xlabel('Strike Price')
plt.ylabel('Implied Volatility Rate')
plt.legend()
plt.show()
```

```python
ref_rates = pd.read_excel('ref_rates.xlsx', sheet_name='data')
```

```python
sofr = ref_rates[['date','SOFR']].ffill()
```

```python
sofr.dropna(inplace=True)
```

```python
sofr.set_index('date', inplace=True)
```

```python
sofr['Rolling Volatility'] = sofr['SOFR'].rolling(window=60).std()
sofr
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(sofr['2025-07-01':].index, sofr['2025-07-01':]['Rolling Volatility'], label='SOFR')
plt.title('SOFR 60-Day Rolling Volatility')
plt.xlabel('Date')
plt.ylabel('Rolling Volatility')
plt.legend()
plt.show()
```

Clearly it is quoted in rates.
