```python
import pandas as pd
import matplotlib.pyplot as plt

rates = pd.read_excel('discount_curve_2025-02-13.xlsx')

def price_bond(face, cpn, freq, T, df: pd.Series):
    periods = int(T * freq)
    coupon_payment = face * cpn / freq
    cash_flows = [coupon_payment] * periods
    cash_flows[-1] += face  # Add face value to the last payment
    #print(cash_flows)
    bond_price = 0
    for i in range( len(cash_flows)):
        discount_factor = df[df['ttm'] == (i + 1) / freq]['discount'].values[0]
        #print(discount_factor)
        bond_price += cash_flows[i] * discount_factor
        
    return bond_price

## Hypothetical bond 1
face = 100
cpn = .0441
freq = 2
T = 5
vanilla_bond1 = price_bond(face, cpn, freq, T, rates[['ttm','discount']])
print(f"Price of the hypothetical vanilla bond 1: {vanilla_bond1 - .196}")

## Hypothetical bond 2
face = 100
cpn = .0441
freq = 2
T = 3
vanilla_bond2 = price_bond(face, cpn, freq, T, rates[['ttm','discount']])
print(f"Price of the hypothetical vanilla bond 2: {vanilla_bond2 - .196}")
```

```python
from scipy.stats import norm
import numpy as np

strike = 100



def black_scholes_call(forward_price, strike, Topt, vol, discount):
    
    d1 = (np.log(forward_price / strike) + 0.5 * vol**2 * Topt) / (vol * np.sqrt(Topt))
    d2 = d1 - vol * np.sqrt(Topt)
    
    call_value = discount * (forward_price * norm.cdf(d1) - strike * norm.cdf(d2))
    return call_value

summed_discount = 0
for i in range(0, 6):
    discount_factor = rates.loc[i]['discount']
    #print(f"Discount factor for year {i}: {discount_factor}")
    summed_discount += discount_factor
#print(f"Summed discount factors for 3 years: {summed_discount}")

vol = .2388*(4.45*.041)
Forward_price = (vanilla_bond1 - summed_discount * cpn * 100 / 2)/(.884084)
print(f"Forward price: {Forward_price}")
```

```python
def calc_ytm(face, cpn, freq, T, price):
    periods = int(T * freq)
    coupon_payment = face * cpn / freq
    cash_flows = [coupon_payment] * periods
    cash_flows[-1] += face  # Add face value to the last payment
    
    def bond_price(ytm):
        discount_factors = [(1 + ytm / freq) ** (-i - 1) for i in range(periods)]
        return sum(cf * df for cf, df in zip(cash_flows, discount_factors))
    
    from scipy.optimize import root_scalar
    result = root_scalar(lambda ytm: bond_price(ytm) - price, bracket=[0, 1], method='bisect')
    
    return result.root

face = 100
cpn = .0441
freq = 2
T = 5
price = vanilla_bond1
ytm = calc_ytm(face, cpn, freq, T, price)
print(f"Yield to Maturity: {ytm:.4%}")
price = vanilla_bond2
ytm = calc_ytm(face, cpn, freq, T, price)
print(f"Yield to Maturity at call date: {ytm:.4%}")
```

```python
from scipy.stats import norm
import numpy as np

strike = 100

def black_scholes_call(forward_price, strike, Topt, vol, discount):
    
    d1 = (np.log(forward_price / strike) + 0.5 * vol**2 * Topt) / (vol * np.sqrt(Topt))
    d2 = d1 - vol * np.sqrt(Topt)
    
    call_value = discount * (forward_price * norm.cdf(d1) - strike * norm.cdf(d2))
    return call_value

summed_discount_plus_1_bp = 0
for i in range(0, 6):
    discount_factor = rates_plus_1_bp.loc[i]['discount']
    #print(f"Discount factor for year {i}: {discount_factor}")
    summed_discount_plus_1_bp += discount_factor
#print(f"Summed discount factors for 3 years: {summed_discount_plus_1_bp}")

vol = .2388*(4.45*.041)
Forward_price_plus_1_bp = (price_plus_1_bp - summed_discount_plus_1_bp * cpn * 100 / 2)/(.884084)
print(f"Forward price plus 1: {Forward_price_plus_1_bp}")


call_price = black_scholes_call(Forward_price_plus_1_bp, strike, 2.95, vol, .884084)
print(f"Call price: {call_price}")
print(f"Price of the callable bond: {price_plus_1_bp- call_price}")

summed_discount_minus_1_bp = 0
for i in range(0, 6):
    discount_factor = rates_minus_1_bp.loc[i]['discount']
    #print(f"Discount factor for year {i}: {discount_factor}")
    summed_discount_minus_1_bp += discount_factor
#print(f"Summed discount factors for 3 years: {summed_discount_minus_1_bp}")    

Forward_price_minus_1_bp = (price_minus_1_bp - summed_discount_minus_1_bp * cpn * 100 / 2)/(.884084)
print(f"Forward price minus 1: {Forward_price_minus_1_bp}")
call_price_minus_1_bp = black_scholes_call(Forward_price_minus_1_bp, strike, 2.95, vol, .884084)
print(f"Call price: {call_price_minus_1_bp}")
print(f"Price of the callable bond: {price_minus_1_bp - call_price_minus_1_bp}")

print(f"Duration of the callable bond: {(price_minus_1_bp - call_price_minus_1_bp - (price_plus_1_bp - call_price)) / (2 * 0.01)}")
```
